```python
import pytri
```

```python
gg = osmnx.graph_from_address("3400 North Charles Street, Baltimore MD")
```

```python
xr = [np.nanmin([n['x'] for _, n in gg.nodes(True)]),np.nanmax([n['x'] for _, n in gg.nodes(True)])]
yr = [np.nanmin([n['y'] for _, n in gg.nodes(True)]),np.nanmax([n['y'] for _, n in gg.nodes(True)])]
```

```python
for i, n in gg.nodes(True):
    gg.node[i]['pos'] = [
        50 * ((n['x'] - xr[0]) / (xr[1] - xr[0])) - 25,
        50 * ((n['y'] - yr[0]) / (yr[1] - yr[0])) - 25,
        0
    ]

gg = gg.to_undirected()

for e in gg.edges(data=True):
    if 'geometry' in e[2]:
        e[2]['geometry'] = 0
```

```python
p = pytri.pytri()
p.show()
```

```python
import networkx as nx
p.graph(nx.Graph(gg))
```
